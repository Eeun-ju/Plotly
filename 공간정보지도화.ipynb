```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel('AML_HOSP.xlsx',index_col='인허가일자',parse_dates=True)
df.head()
```

```python
fig = px.scatter_mapbox(df,
                        lat="ycoord", # latitude 위도
                        lon="xcoord", # longitude 경도
                        hover_name="사업장명", # 가장 위에 오는 정보 이름
                        hover_data=["소재지전체", "도로명전체"], # 정보 데이터 표시
                        color_discrete_sequence=["gray"], # 색상 표시 ex) lightyellow, lime, limegreen, linen, magenta, maroon
                        zoom=5,# ex) 지도 확대 값
                        height=400)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
